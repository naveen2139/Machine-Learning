```python
ships=conn["sample_geospatial"]
shipcoll=ships["shipwrecks"]
c=shipcoll.find()
shiplist=list()
for i in c:
  shiplist.append(i)
  # print(i)
shipdf=pd.DataFrame(shiplist)
shipdf
```
